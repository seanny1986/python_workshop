# Linear Regression

## 1. Introduction

Regression is the art of learning a function that fits a set of known datapoints, and is an integral component of modern machine learning. In a nutshell, we have a dataset $\mathcal{D}$, consisting of known inputs and outputs $(\mathbf{X}, \mathbf{Y})$, where $\mathbf{X}$ is a vector of inputs, and $\mathbf{Y}$ is a vector of outputs. In regression, we learn a function or model $f : \mathbf{X} \rightarrow \mathbf{Y^{*}}$, where $\mathbf{Y^{*}}\approx \mathbf{Y}$.

To illustrate the basic idea and motivation, let's say we have the following set of datapoints:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)                         # x range
A = np.random.randn()*10                            # generate random slope
B = np.random.randn()*10                            # generate random intercept
y = (A*x+B)+1.2*np.random.randn(100)                # noisy data
y_actual = (A*x+B)
plt.scatter(x,y)
plt.plot(x, y_actual, "--r")
plt.title("Simulated Noisy Dataset: Wealth vs. Happiness")
plt.xlabel("Wealth")
plt.ylabel("Happiness")
plt.legend(["actual", "noisy"])
plt.show()
```

Let's imagine that this data captures some underlying relationship, such as Wealth versus Happiness, and that there's a clear relationship between the two variables. That is, we can say that as wealth increases, happiness increases (or decreases) and vice versa. Having a good idea of the underlying relationship between the two variables is useful. Furthermore, what if we could somehow learn a predictive function such that -- if someone gave us a given wealth value -- we could estimate their happiness based on empirical data? That seems like it could be a neat and powerful idea!

So how could we go about doing this?

Intuitively, a human could look at the points and guess that there's an underlying linear relationship here -- and in fact, I've plotted the actual relationship that I used to generate the data with the dashed red line. Assuming that we didn't already *know* the underlying relationship, we might assume that the relationship is linear, and fit a line of the form $y=mx+c$ to the data to get a rough predictive model for any input x. The question then becomes -- *which* line goes through these points, and how do we find it?

Knowing the equation of a line, regression is the process of trying to find the function that best fits the datapoints. We can rewrite the equation for a line as $y = \mathbf{\Theta^{T} X}$, where $\mathbf{\Theta} = [\theta_{0}; \theta_{1}]$, and $\mathbf{X} = [x_{0}; x_{1}]$, with $x_{0} = 1$. We can now reframe the task of finding the right line as the task of finding the right parameter weights $\mathbf{\Theta}$, such that $Y^{*} = \mathbf{\Theta^T X}$, and $\mathbf{Y^{*}} \approx \mathbf{Y}$. This task of finding the best parameter weights is the central focus of parametric regression.

So how do we find these parameter weights? We use optimization.


## 2. Optimization

Mathematical optimization is the task of maximizing or minimizing a function $f: X \rightarrow \mathbb{R}$ by changing the input to the function (note that this $X$ is different to the X of the previous section. Here, $X$ is the argument to our cost function, whereas in the previous section, $X$ and $\theta$ are inputs to our regression function that gives us $Y^{*}$). That is, we want to find the value(s) of $X$ that result in maxima or minima:

![alt text](img/maxima.png)

There are many ways to do this. Looking at the function above, we can see the the derivative is zero at the maxima and local minima; if we have a simple function, we could take the derivative, and directly solve for the points where the derivative is zero. Then, going back to our original function, we could determine if we found a maximum or minimum. However, this strategy quickly breaks down in situations where our function is non-differentiable, or where we have no way of solving for the zeros in closed form. We can get around this using another strategy: maybe we just randomly vary the input and store the best value in memory? That would work, but it breaks down in large parameter spaces. Another strategy might be to use calculus, but instead of trying to solve for the points at which the derivative function is zero, we just take a small step in that direction. 

These are just a few out of many, many principled techniques used for optimization, with some being inspired by nature (evolutionary strategies, simulated annealing, particle swarm optimization, ant-colony optimization, etc.), and others being a clever application of mathematical principles learned in high school (for example, calculus-based methods).

When doing regression, we want to find the best set of parameter weights given some set of data, which we can do by building a **cost function** $f: \mathcal{X}\times\mathbf{\Theta} \rightarrow \mathbb{R}$ for a set of parameters $\theta \in \Theta$ that tells us how good the weights are on average. From there, we can use a standard optimization method to find the best set of weights. For regression problems, the score function we typically use is the mean-squared error:

\begin{equation}
\mathcal{L} = \frac{1}{2N}\sum_{i=1}^{N}-(y(x_i, \theta)-y_{i}^{*})^{2}
\end{equation}

where $y_{i}^{*}$ is a known datapoint, and $y(x_i, \theta)$ is:

\begin{equation}
\mathbf{Y} = \mathbf{\theta^{T} X}
\end{equation}

Plotting this function:

```python
import numpy as np
import matplotlib.pyplot as plt

L = lambda x : -(x-mu)**2
C = lambda x : (x-mu)**2
mu = 0
xs = np.linspace(-5, 5, 100)
plt.plot(xs, L(xs))
plt.plot(xs, C(xs))
plt.scatter(0,0, color="r")
plt.title("Cost Function Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["cost function", "negative cost function", "maxima / minima"])
plt.show()
```

We can see that it has a single defined maximum at $y=y^{*}$ (in this case, I've called them x and $\mu$ for reasons that will become clearer later). Furthermore, we can see that maximizing the cost function $\mathcal{L}$ is the same as *minimizing* the negative cost function $\mathcal{C}$ -- in this case, given by the orange line.

### 2.1 An Example

Below, we implement a cost function that uses our parameter weights, and calculates the average loss:

```python
def cost(theta, x, targ):
    pred = theta.T.dot(x)
    loss = -0.5*(pred-targ)**2
    loss = -np.mean(loss, axis=-1)
    return loss
```

Our cost function takes in a prediction and a target, and returns a scalar value that can take a maximum value of zero. We take the mean over whatever the final axis is so that we can also use this cost function with population-based optimization methods (evolutionary strategies). Since we're using a minimization function, we minimize the negative loss, which -- as we showed above -- is equivalent to maximizing the positive loss function:

$\max : f(X;\theta) \equiv \min : -f(X;\theta)$

Next, we'll pass this cost function to an inbuilt optimizer, and find the optimum $\mathbf{\theta}$.

```python
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

samples = 1000
x = np.linspace(-1, 1, samples)                         # x range
A = np.random.randn() * 10                            # generate random slope
B = np.random.randn() * 10                            # generate random intercept
y = (A * x + B) + 2 * np.random.randn(samples)                # add noise to the function

x_vec = np.vstack([x, np.ones(x.shape)])
theta_0 = np.random.randn(x_vec.shape[0], 1)
theta = minimize(cost, theta_0, args=(x_vec, y), method="CG")

print("A value: ", A)
print("B value: ", B)
print("Found values: ", theta.x)

y_pred = np.array(theta.x).dot(x_vec)
plt.figure(figsize=(10,10))
plt.scatter(x, y)
plt.plot(x, y_pred, "r")
plt.show()
```

As we can see, the optimized parameters match the actual weighting values very closely! Plotting the line given by these weights, we can see that it passes directly through the datapoints.

So what did we just do? This is an example of **Linear Regression**, where we used a linear relationship between a set of parameter weights, and an assumed underlying relationship in the data to fit a function that broadly approximates the underlying relationship.

In this case, we used a linear **basis function**, but we could also have used a polynomial basis function as well if we thought it fit the data, for example $\mathbf{X} = [1; x; x^2; ...; x^n]$. Other common choices include splines and Gaussian basis functions.

### 2.2 Interpretation Using Information Theory
What we've just done is an example of **Maximum Likelihood Estimation** (MLE). Most of the score functions that we typically use have a probabilistic interpretation, where we maximize the log-probability of the weights given the data. Since $\log{p(x)}$ and $p(x)$ are maximized at the mean, and have the same mean, it's often easier to maximize the log probability function since it turns messy products into sums. In addition, it can also help avoid numerical errors such as exploding or vanishing gradients that can result as consequence of the exponential function.

In the case of the mean squared-error function, we are maximizing the log-probability of a normal distribution, since:

\begin{equation}
P(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}
\end{equation}

Then:

\begin{equation}
\log{P(x)} = -\frac{(x-\mu)^{2}}{2\sigma^{2}}\log{\frac{1}{\sigma\sqrt{2\pi}}} = k(x-\mu)^{2}
\end{equation}

It is common for cost functions to include additional terms, which can be motivated under this framework. For example:

\begin{equation}
\log{P(x_{0}, x_{1}, x_{2},...,x_{N})} = \sum_{i=0}^{N} k_{i}(x_{i}-\mu_{i})^{2}
\end{equation}

In which we maximize the probability of several criteria simultaneously, with weightings for each. Similarly, we can also use Bayes' Theorem to incorporate prior evidence into our regression function. Bayes' theorem involves finding a posterior *distribution* using the likelihood, a prior, and a normalizing constant. Since $P(A|B)P(B)=P(B|A)P(A)$, this is expressed more formally as:

$P(B|A) = \frac{P(A|B)P(B)}{P(A)}$

Or in log form, using $\theta$ and $y$:

\begin{equation}
\log{P(\theta|y)} = \log{P(y|\theta)}+\log{P(\theta)}-\log{P(\mathcal{y})}
\end{equation}

The term $P(B|A)$ is called the posterior, and represents the probability of an outcome -- called the *likelihood* or *evidence*, represented by $P(A|B)$ -- that is weighted by a belief about given information called a *prior* -- in this case $P(A)$. The term $P(B)$ is a normalizing constant, so it's quite common to shorten Bayes' theorem to state that $P(B|A) \propto P(A, B)$. Typically, finding the posterior distribution is a difficult process (that has certain advantages), so it's more common to find the mean of the posterior distribution, which is known as finding the **Maximum A Priori**. One way that this is commonly done is in the form of **regularization**, where we place a prior on the parameter weights $\theta$ of our regression model. For example, we might assume that the weights should be $\theta \sim \mathcal{N}(0,\mathbb{I})$, which would favor simpler models. The cost function for this regression problem can be expressed as:

\begin{equation}
\log{P(\theta|y)} \approx -k_{1}(f(\theta)-y^{*})^{2}-k_{2}\theta^{2}+C
\end{equation}

This would be called $L2$ regularization, as opposed to $L1$ regularization in which the weights are penalized according to the absolute value. Weight regularization is a commonly used trick for tuning models, as it helps prevent them from overfitting to the data.

To demonstrate some of these points, I've plotted the log-likelihood below, along with the Gaussian function:

```python
from math import sqrt, pi

normal = lambda x, mean, std : 1/std/sqrt(2*pi)*np.exp(-0.5*((x-mean)/std)**2)

mu = 0
stdev = 1
xs = np.linspace(-2, 2, 100)
probs = normal(xs, mu, stdev)
logprobs = np.log(probs)

# normalize and translate
probs /= np.max(probs)
logprobs += np.max(np.abs(logprobs))
logprobs /= np.max(logprobs)

# plot
plt.plot(xs, probs)
plt.plot(xs, logprobs)
plt.plot(np.zeros(xs.shape), np.linspace(0,1,100), "--r")
plt.title("Gaussian Function vs. Log Gaussian Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["Gaussian Function", "Log Gaussian Function", "Mean"])
plt.show()
```

I've normalized the probability and log-probability functions above so that they sit on top of one another. As you can see from the plot, both functions have the same mean and maximum.

```python
from scipy.optimize import minimize

def regularized_cost(theta, x, targ, w):
    pred = theta.T.dot(x) + w * theta.T.dot(theta)
    loss = -0.5 * (pred-targ) ** 2
    loss = -np.mean(loss, axis=-1)
    return loss

x = np.linspace(-1, 1, 100)                         # x range
A = np.random.randn() * 10                            # generate random slope
B = np.random.randn() * 10                            # generate random intercept
y = (A * x + B) + 2 * np.random.randn(100)                # add noise to the function

data = []
ws = np.linspace(0, 1, 5)
for w in ws: 
    x_vec = np.vstack([x, np.ones(x.shape)])
    theta_0 = np.random.randn(x_vec.shape[0],1)
    theta = minimize(regularized_cost, theta_0, args=(x_vec, y, w), method="CG")
    data.append(theta.x)

plt.figure(figsize=(20,4))
for i, (d, w) in enumerate(zip(data, ws.tolist())):
    y_pred = np.array(d).dot(x_vec)
    plt.subplot(1, len(data), i + 1)
    plt.scatter(x, y)
    plt.plot(x, y_pred, "r")
    plt.title("W="+str(w))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(["fitted function", "data"])
plt.show()
```

As we can see, the regularization parameter has a linearizing effect on the prediction. This is detrimental in this case, because we have a lot of tightly clustered datapoints, but can be quite useful in other problems (for example, regression).

In general, the type of loss function that you use will depend on the type of output that you expect. For continuous outputs, you would typically use a mean squared-error function; for categorical problems using the logit function (which we will cover later) you would typically use a binary / categorical cross entropy loss function (these are the log-probability functions of the binomial and multinomial distributions, respectively). For the softmax output, we use the cross-entropy loss. When applying regularization, a common technique is to do a hyperparameter search to determine the best hyperparameter settings by comparing accuracy on the training set.

### 2.2 Gradient Descent and Evolutionary Strategies
In the previous section, we used a standard optimizer from scipy to do the heavy lifting for us. In this section, we'll briefly cover two simple optimization methods: gradient descent, and evolutionary strategies. Gradient descent is the workhorse of deep learning, in which we train deep neural networks. Evolutionary strategies are conceptually simple and embarrassingly effective optimizers that make for a good sanity check against more complex methods. Their primary weakness is that they don't scale well to large parameter spaces, limiting their usefulness in applications like deep learning where a single network can contain millions of parameters.

Gradient descent is conceptually straightforward: we find the gradient of our function $f$, and then we take a small step in the direction that maximizes or minimizes the function (whatever our goal is):

![title](img/gradient_descent.jpg)

In general, for a function $f(x)$, we can step down the direction of steepest descent such that $f(x_{1}) = f(x_{0}-\delta)$, where $\delta = \alpha\frac{\partial}{\partial x} J(x_{0})$. We've implemented a basic gradient descent function below:

```python
def gradient_descent(fun, theta, args, max_iters=100, thresh=1, alpha=1e-3, log_freq=10):
    iteration = 1
    while True:
        loss, grad = cost(theta, args[0], args[1])
        if iteration % log_freq == 0:
            print("--- Iteration: {}, Loss: {:.5f} ---".format(iteration, loss))
        theta += alpha*(args[0]).dot(grad.T)
        if abs(loss) < thresh:
            break
        if iteration > max_iters:
            break
        iteration += 1
    return theta
        
def cost(theta, x, targ):
    pred = theta.T.dot(x)
    err = pred-targ
    loss = -0.5*err**2
    loss = -np.mean(loss)
    grad = -err
    return loss, grad

samples = 1000
x = np.linspace(-1, 1, samples)                         # x range
A = np.random.randn() * 10                            # generate random slope
B = np.random.randn() * 10                            # generate random intercept
y = (A * x + B) + 2 * np.random.randn(samples)                # add noise to the function

x_vec = np.vstack([x, np.ones(x.shape)])
theta_0 = np.random.randn(x_vec.shape[0],1)
theta = gradient_descent(cost, theta_0, args=(x_vec, y))

print("A value: ", A)
print("B value: ", B)
print("Found values: ", theta[:,0])

y_pred = np.array(theta).T.dot(x_vec)
plt.figure(figsize=(10,10))                                  # create a figure
plt.scatter(x,y)
plt.plot(x, y_pred[0,:], "r")
plt.title("Regression Using First Order Gradient Descent")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["fitted function", "data"])
plt.show()
```

As we can see, we've managed to fit the function nicely.

The advantage of this method is that it is straightforward to understand (providing you have a grasp of calculus) and it is easy to implemnt. It can also be very effective -- neural networks, for example, are typically optimized using variants of stochastic gradient descent. The downside of the method is that it can get trapped in saddle-points if the loss surface is non-convex. In addition, it relies on a fixed step-size, which can affect the convergence of the algorithm. A large step-size can cause it to blow up, whilst a small step-size takes a very long time to optimize.

These problems two problems can be solved using second-order methods that take the second derivative into account. An intuitive explanation is that they take curvature or acceleration into account, and so their step-size goes to zero as they approach a minima. Additionally, because these methods take curvature into account, they are able to escape saddle points, and are guaranteed to find at least a local minima. The downside of these methods is that they often involve a costly inversion of a Hessian matrix (i.e. a matrix of second derivatives). If this matrix is ill-conditioned (say, our parameters are over-specified), then we are likely to run into difficulties during optimization. As a result, these methods are generally used for small parameter spaces.

A happy medium are first order methods that approximate the second derivative using a momentum term. This momentum term lets them escape saddle points and local optima, but they generally aren't as efficient as true second-order methods.

### 2.4 Evolutionary Strategies
Another strategy for minimizing a cost function is to use an evolutionary strategy. Mathematically, both gradient descent and evolutionary strategies are still doing the same thing, as they are both maximizing the log probability of some set of parameters given a dataset. However, evolutionary strategies are very straightforward to implement, and are often embarassingly effective. In a nutshell, the strategy is:

1. We initialize a mean and standard deviation for each parameter;
2. We sample a set of $n$ parameters using the means and standard deviations;
3. We evaluate all of the parameters we've sampled using the cost function;
4. We rank all of the parameters by their cost;
5. We select the top 20% (or so) of the population, and calculate new means and standard deviations based on their parameters.

This method, known as the cross-entropy method, is gradient free, which is advantageous since it doesn't require a function to be differentiable in order to work. The downside is that gradient-based methods may outperform it in large parameter spaces. This method is first order, though second-order methods like Covariance Matrix Adaptation rely on a similar idea, and can be very effective.

To demonstrate the idea, we'll use the method to fit a line to a noisy dataset drawn from a quadratic function. First, we'll implement the optimization routine as outlined above:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def optimize_f(x, y_targ, mean=0., std=100., samples=10000, elite_frac=0.2, deg=1, max_iters=100, thresh=1e-2, log_freq=10):
    # outer optimization function
    
    def cost_func(pred, targ):
        # mean squared-error cost function
        loss = -0.5*(pred-targ)**2
        return np.mean(loss, axis=-1)
    
    x = np.vstack([x**i for i in reversed(range(deg+1))])
    best_n = int(elite_frac*samples)
    iteration = 1
    prev_cost = None
    delta = None
    mu, dev = np.ones((deg+1,))*mean, np.ones((deg+1,))*std
    running = True
    while running:
        theta = np.random.normal(mu, dev, (samples, deg+1))
        y_pred = theta.dot(x)
        loss = cost_func(y_pred, np.tile(y_targ, (samples, 1)))
        idxs = np.argsort(loss)[::-1]
        best = idxs[0:best_n]
        absolute_cost = np.abs(loss[idxs[best]].mean())
        if iteration % log_freq == 0:
            print("--- Iteration: {}, Loss: {:.5f} ---".format(iteration, loss[idxs[best]].mean()))
        if absolute_cost <= thresh:
            print("Termination threshold reached. Current cost is {:.5f}".format(absolute_cost))
            running = False
        elif iteration >= max_iters:
            print("Maximum iterations reached. Current cost is {:.5f}".format(absolute_cost))
            running = False
        elif prev_cost is not None:
            delta = abs(absolute_cost-prev_cost)
            if delta <= 1e-5:
                print("Algorithm converged after {} iterations at cost {:.5f}".format(iteration, absolute_cost))
                running = False
        best_theta = theta[best, :]
        mu, dev = best_theta.mean(axis=0), best_theta.std(axis=0)
        prev_cost = absolute_cost
        iteration += 1
    return mu
```

This is a bit to take in, so we'll go through it line-by-line:
* Lines 4-7 are our mean squared-error cost function. Our input to this function is an $M\times N$ matrix, where $M$ is the number of samples in our population, and $N$ is the number of datapoints. We take the mean over the $N$ axis, and return an $M \times 1$ vector of scores for each member of the population.
* On line 9, since the function is quadratic, so too must our features be. For a second-degree polynomial, our range needs to be $\{x^i, \forall i \in 2, 1, 0\}$ .
* Our means stdevs are initialized on line 14 as [0,0,0] and [100, 100, 100]
* We sample thetas on line 17, and make a prediction on line 18. We evaluate these predictions with the aforementioned loss function on line 19.
* On line 20, we sort the scores in descending order. We have to get the indexes from the array, since these correspond to the population member.
* On line 21, we grab the indexes of the best N members of the population (in this case, the top 20%). We use these indexes to grab the best scoring parameters on line 35.
* We combine the parameters of the best N members by taking the mean and standard deviation for each parameter, and setting that as the new mean and stdev.

Below, we run it to test it out:

```python
samples = 1000
x = np.linspace(-1, 1, 1000)                         # x range
A = np.random.randn()*10                            # generate random slope
B = np.random.randn()*10                            # generate random intercept
C = np.random.randn()*10
y = (A * x ** 2 + B * x + C) + 4 * np.random.randn(1000)        # add noise to the function

theta = optimize_f(x, y, deg=2)           # call optimization fn
print("A: {}".format(A))
print("B: {}".format(B))
print("C: {}".format(C))
print("Optimized Theta: {}".format(theta))
x_vec = np.vstack([x**i for i in reversed(range(3))])
y_pred = theta.reshape(1,-1).dot(x_vec)

plt.figure(figsize=(10,10))
plt.scatter(x, y)
plt.plot(x, y_pred[0,:], "r")                 # plot the line given by THETA
plt.title("Regression Using Evolutionary Strategy (CEM)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["fitted function", "data"])
plt.show()   
```

As we can see, we managed to fit the function quite nicely. In this case, we used quadratic basis function, since we had some idea that the data follows a quadratic distribution.

### 2.3 Convexity of the Loss Function

An important note to point out: in convex optimization, our loss surface is convex, meaning that there is a single global maxima or minima, and no local maxima or minima. Think a quadratic function. These are nice problems to work with, because it can be guaranteed that our function will converge to the global minimum. 

In practice, most interesting problems have a non-convex loss surface, meaning that they can be fraught with local minima. One way to think about this concept is to imagine a ball rolling down a hill. Depending on the shape of the hill, the ball can get stuck at a local minima or flat region when a lower minima exists. As a result, our function may never reach the global minima. In the case of neural networks, this is certainly the case -- the chances of finding a global minima are effectively zero, though in practice this may not matter.

To demonstrate some of these ideas, see the following example of a non-convex surface. If you needed an algorithm to navigate to the global minimum, what would you use? In general, this is a very difficult problem with no best solution.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

def f(xy):
    pdf = w1 * d1.pdf(xy) + w2 * d2.pdf(xy) - w3 * d3.pdf(xy)
    return -z * pdf

z = 0.1
w1, w2, w3 = 0.4, 0.35, 0.35
d1 = multivariate_normal([2, -2], [[0.5, 0.3], [0.3, 0.5]])
d2 = multivariate_normal([-0.5, 0], [[0.5, 0.3], [0.3, 0.5]])
d3 = multivariate_normal([-3, -3], [[1., 0.], [0., 1.]])

size = 100
xs = np.linspace(-5, 5, size)
ys = np.linspace(-5, 5, size)
X, Y = np.meshgrid(xs, ys)
vals = np.array([[f(np.array([X[i,j], Y[i,j]])) for j in range(size)] for i in range(size)])

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, vals)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

If we used a gradient descent-based algorithm on this surface, depending on our initialization point, we may end up trapped in a loal optima or on a plateau.

Interestingly, this is an issue in many classical optimization problems, but doesn't seem to be an issue in deep learning. Even with first-order optimizers that get trapped in local optima (in fact, the chances of finding the global optima are effectively zero when training a neural network), the current generation of neural network based machine learning methods are still able to learn functions of the data that capture complex relationships beyond more traditional methods, and that typically generalize well to new data. Why this is the case is still an active area of research.

## 3. Neural Network Regression
Now that we've covered these basics, we're equipped to look at neural nets, and more modern machine learning techniques. The previous regression examples relied on us having some kind of human intuition as to the degree of the function we would fit, but what if the function isn't a polynomial, or even worse, is so complex that it would be impossible for us to guess a function that would regress accurately? One solution to this problem is to use neural networks.

Neural networks take inspiration from the networks of neurons in the brains of biological organisms.

Each neuron takes input from a series of other neurons, does a non-linear transformation on this input using an **activation function**, and then passes this output on to a series of connected neurons. The large number of parameters, and the non-linearity of the activation functions means that neural nets can approximate extraordinarily complex functions that would be difficult or impossible to represent any other way. Like clusters of neurons in biological brains, artificial neural networks are trained by altering the strength of the connections between neurons (the parameter weights $\Theta$). In practice, this is no different to what we've done earlier; we can use both evolutionary algorithms and gradient descent to train our networks, though gradient-based methods typically scale far better (evolutionary algorithms struggle in very large parameter spaces).

### 3.1 The Backpropagation Algorithm

The backpropagation algorithm (or just backprop) is the workhorse of deep learning, and is a combination of gradient descent with the chain rule from calculus. For the following network:

![title](img/neural_net.png)

We have a single hidden layer, and two layers of weights that need to be trained. For each weight, our goal is to find $\frac{\partial \mathcal{L}}{\partial \Theta}$, where $\mathcal{L}$ is our cost or loss function. This is tricky due to the distributed nature of the network; the effect that each individual neuron has on the error is split among the neurons it feeds into.

If we feed forward through the network:

\begin{equation}
y = f_{n}(\Theta_{n}^{T}f_{n-1}(\Theta_{n-1}^{T}f_{n-1}(...)))
\end{equation}

\begin{equation}
\mathcal{L} = -\frac{1}{2}(y-y^{*})^2 = -\frac{1}{2}(f_{n}(\Theta_{n}^{T}f_{n-1}(\Theta_{n-1}^{T}f_{n-2}(...)))-y^{*})^2
\end{equation}

Letting:

\begin{equation}
z_{i} = \Theta_{i}^{T}f_{i-1}(z_{i-1})
\end{equation}

Then:

\begin{equation}
\frac{\partial \mathcal{L}}{\partial \Theta_{n}} = \frac{\partial \mathcal{L}}{\partial y}\frac{\partial y}{\partial z_{n}} \frac{\partial z_{n}}{\partial \Theta_{n}}
\end{equation}

The derivatives on the left and right are straightforward to calculate:

\begin{equation}
\frac{\partial \mathcal{L}}{\partial y} = -(y-y^{*})
\end{equation}

\begin{equation}
\frac{\partial z_{n}}{\partial \Theta_{n}} = f(z_{n-1})
\end{equation}

The middle term requires that our activation functions be differentiable in order to push a gradient back through the network, since:

\begin{equation}
\frac{\partial y}{\partial z_{n-m}} = \frac{\partial y}{\partial z_{n}}\prod_{i=n}^{n-m+1}\frac{\partial z_{i}}{\partial z_{i-1}}
\end{equation}

Where:

\begin{equation}
\frac{\partial z_{i}}{\partial z_{i-1}} = \Theta_{i}^{T}f'(z_{i-1})
\end{equation}

In this way, we can successively calculate that gradients of the weights at each layer by first calculating the loss, taking it's derivative, and then pushing back through the network layer by layer. The advantages of this algorithm are that it scales well, and it's general -- aside from requiring differentiable functions, we can use any activation functions that we like.

Once we have the gradients, we can update the network using any gradient-based optimization method of our choice. Common methods include stochastic gradient descent (gradient descent using labels sampled randomly), RMSprop, and Adam, which add a momentum term to help escape local minima.

### 3.2 An Implementation
The following is a short Multi-Layer Perceptron (MLP -- a type of feedforward neural network) implemented in PyTorch:

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)   # hidden layer
        self.predict = torch.nn.Linear(n_hidden, n_output)   # output layer

    def forward(self, x):
        x = F.elu(self.hidden(x))      # activation function for hidden layer
        x = self.predict(x)             # linear output
        return x
```

We inherit from Torch's nn Module, which implements all of the linear algebra for us. We use torch's Linear layers to represent the fully connected layers within our network. This is a single-layer network, corresponding to two weight matrices between our neurons.

Next, we'll run our training loop on a set of noisy data, as we did above.

```python
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)  # x data (tensor), shape=(100, 1)
y = x.pow(2) + 0.2*torch.rand(x.size())                 # noisy y data (tensor), shape=(100, 1)
net = Net(n_feature=1, n_hidden=32, n_output=1)
optimizer = torch.optim.SGD(net.parameters(), lr=1e-1)
loss_func = torch.nn.MSELoss()

for t in range(500):
    prediction = net(x)     # input x and predict based on x
    loss = loss_func(prediction, y)     # must be (1. nn output, 2. target)
    optimizer.zero_grad()   # clear gradients for next train
    loss.backward()         # backpropagation, compute gradients
    optimizer.step()        # apply gradients

print("Final loss is: {:.5f}".format(loss.item()))
plt.scatter(x.data.numpy(), y.data.numpy())
plt.plot(x.data.numpy(), prediction.data.numpy(), 'r-')
plt.show()
```

As we can see, the network learns to approximate the dataset. However, it seems that the fit isn't as good as when we used polynomial regression to fit this type of function. Why is that? We can get an idea by seeing what happens if we get the network to make predictions beyond the domain of its training set.

This type of network is called a Multi-Layer Perceptron, and is one of the simplest types of neural net.

```python
x = torch.unsqueeze(torch.linspace(-5, 5, 100), dim=1)
prediction = net(x)
plt.plot(x.data.numpy(), prediction.data.numpy(), 'r-')
plt.show()
```

As we can see, the network is linear outside of the where it has been trained on the dataset.

## 4. Time Series Regression Using Recurrent Neural Nets

Recurrent Neural Nets (RNNs) are an architecture that is designed for time-series data. A standard neural net can be rolled out through time such that $y_{t_{N}} = f(f(f(...f(y_{t_{N-t+1}}))))$. When we backpropagate through this, we accumulate a very long chain of derivatives, which can result in exploding or vanishing gradients. To get around this, we typically use Long Short-Term Memory (LSTM) units, or Gated Recurrent Units (GRU) to push derivatives back over long time periods. Compared to standard networks, these are functions $y_{t_{N}} = f(f(f(...f(y_{t_{N-t+1}}, h(x_{N-t})))))$, where $h$ denotes the hidden units in the network. This gives them an internal memory that is useful in other applications, and lets them "skip" derivatives, increasing the distance we can backpropagate through time. 

In the following example, we will train an RNN to fit some time simple time series data, and compare it to a standard MLP.

```python
class RNN(torch.nn.Module):
    def __init__(self):
        super(RNN, self).__init__()

        self.rnn = torch.nn.RNN(
            input_size=INPUT_SIZE,
            hidden_size=32,     # rnn hidden unit
            num_layers=1,       # number of rnn layer
            batch_first=True,   # input & output will has batch size as 1s dimension. e.g. (batch, time_step, input_size)
        )
        self.out = torch.nn.Linear(32, 1)

    def forward(self, x, h_state):
        # x (batch, time_step, input_size)
        # h_state (n_layers, batch, hidden_size)
        # r_out (batch, time_step, hidden_size)
        r_out, h_state = self.rnn(x, h_state)
        outs = []    # save all predictions
        for time_step in range(r_out.size(1)):    # calculate output for each time step
            outs.append(self.out(r_out[:, time_step, :]))
        return torch.stack(outs, dim=1), h_state
```

Next we'll run the training loop:

```python
import numpy as np

# Hyper Parameters
TIME_STEP = 10      # rnn time step
INPUT_SIZE = 1      # rnn input size
LR = 0.02           # learning rate

rnn = RNN()
print(rnn)

optimizer = torch.optim.Adam(rnn.parameters(), lr=LR)   # optimize all cnn parameters
loss_func = torch.nn.MSELoss()

h_state = None      # for initial hidden state

plt.figure(1, figsize=(12, 5))

for step in range(100):
    start, end = step * np.pi, (step+1)*np.pi   # time range
    # use sin predicts cos
    steps = np.linspace(start, end, TIME_STEP, dtype=np.float32, endpoint=False)  # float32 for converting torch FloatTensor
    x_np = np.sin(steps)
    y_np = np.cos(steps)

    x = torch.from_numpy(x_np[np.newaxis, :, np.newaxis])    # shape (batch, time_step, input_size)
    y = torch.from_numpy(y_np[np.newaxis, :, np.newaxis])

    prediction, h_state = rnn(x, h_state)   # rnn output
    # !! next step is important !!
    h_state = h_state.data        # repack the hidden state, break the connection from last iteration

    loss = loss_func(prediction, y)         # calculate loss
    optimizer.zero_grad()                   # clear gradients for this training step
    loss.backward()                         # backpropagation, compute gradients
    optimizer.step()                        # apply gradients

# plotting
plt.plot(steps, y_np.flatten(), 'r-')
plt.plot(steps, prediction.data.numpy().flatten(), 'b-')
plt.show()
```

As we can see, the RNN learns to predict the function output. As an exercise, I encourage you to try applying an MLP to the task of learning a cyclical function like $\sin(x)$ or $\cos(x)$ to see how it performs. RNNs are useful in situations where features can be important over long time horizons, and in which we would like to learn those features directly.
